programa que solicite ao usuário dois números e exiba a soma, subtração, multiplicação, divisão, divisão inteira, módulo e exponenciação desses números.

```python
num1 = float(input("Digite o primeiro número: ")) #criação da variavel1 e recebimento do valor pelo usuário
num2 = float(input("Digite o segundo número: ")) #criação da variavel2 e recebimento do valor pelo usuário
Dvd = num1 / num2 #criação da variavel para obter o valor da divisão

print(f"Soma: {num1 + num2}")
print(f"Subtração: {num1 - num2}")
print(f"Multiplicação: {num1 * num2}")
print(f"Divisão: {Dvd:.2f}") # "Dvd:.2f" para puxar o valor e limitar 2 números pós a virgula
print(f"Divisão inteira: {num1 // num2}")
print(f"Módulo: {num1 % num2}")
print(f"Potência: {num1 ** num2}")
```

programa que solicite ao usuário uma nota (0-10) e exiba uma mensagem indicando se ele foi aprovado (nota >= 7), está de recuperação (nota entre 5 e 6.9) ou foi reprovado (nota < 5).


```python
nota = float(input("Digite sua nota: "))#criação da variavel1 e recebimento do valor pelo usuário

if nota >= 7:
  print("Você foi aprovado, Parabens!!!")
elif nota >= 5:
    print("Você está de recuperação |:")
else:
    print("Você está reprovado :(")
```
